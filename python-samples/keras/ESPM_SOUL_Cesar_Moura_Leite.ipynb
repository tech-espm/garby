```
#Creating classes array for classification
class_names=["Cardboard", "Glass", "Metal", "Paper", "Plastic", "Trash"]
```

```
#Loading single image and importing used libraries
from numpy import asarray
from os import listdir
from matplotlib import image, pyplot
import pandas as pd
import numpy as np

img = image.imread('Garbage classification/cardboard/cardboard1.jpg')
pyplot.imshow(img)
```

# Dataset Consrtuction

Creating list that will hold both the images and its classifications so that we can concatenate them on a Pandas Dataframe later

```
loaded_images = list()
image_labels = list()
```

Loding Cardboard Images

```
for filename in listdir('Garbage classification/cardboard'):
  img_data = image.imread('Garbage classification/cardboard/' + filename)
  image_labels.append(class_names[0])
  loaded_images.append(asarray(img_data))
```

Loading Glass Images

```
for filename in listdir('Garbage classification/glass'):
  img_data = image.imread('Garbage classification/glass/' + filename)
  image_labels.append(class_names[1])
  loaded_images.append(asarray(img_data))
```

Loading Metal Images

```
for filename in listdir('Garbage classification/metal'):
  img_data = image.imread('Garbage classification/metal/' + filename)
  image_labels.append(class_names[2])
  loaded_images.append(asarray(img_data))
```

Loading Paper Images

```
for filename in listdir('Garbage classification/paper'):
  img_data = image.imread('Garbage classification/paper/' + filename)
  image_labels.append(class_names[3])
  loaded_images.append(asarray(img_data))
```

Loading Plastic Images

```
for filename in listdir('Garbage classification/plastic'):
  img_data = image.imread('Garbage classification/plastic/' + filename)
  image_labels.append(class_names[4])
  loaded_images.append(asarray(img_data))
```

Loading Trash Images

```
for filename in listdir('Garbage classification/trash'):
  img_data = image.imread('Garbage classification/trash/' + filename)
  image_labels.append(class_names[5])
  loaded_images.append(asarray(img_data))
```

Displaying Final List Sizes

```
print(len(loaded_images))
print(len(image_labels))
print(type(loaded_images[0]))
```
